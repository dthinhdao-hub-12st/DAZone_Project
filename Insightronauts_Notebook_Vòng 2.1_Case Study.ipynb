## **HANDLING NULL ENTRIES**

```python
log_numerical_data[log_numerical_data['brand_id'].isna()].sort_values(by = 'item_id')
```

Idea: Every item should be linked with a brand, if the item with null brand has been recorded before, may be the brand id can be recollect. Count how many distinct items with the missing brand has already appear with the not missing brand

```python
log_numerical_data.loc[log_numerical_data['brand_id'].isna(), 'item_id'] \
    .isin(log_numerical_data.loc[log_numerical_data['brand_id'].notna(), 'item_id']) \
    .groupby(log_numerical_data.loc[log_numerical_data['brand_id'].isna(), 'item_id']) \
    .any().sum()
```

```python
item_brand_map = log_numerical_data.dropna(subset=['brand_id']) \
                                   .drop_duplicates(subset=['item_id']) \
                                   .set_index('item_id')['brand_id']

log_numerical_data['brand_id'] = log_numerical_data['brand_id'].fillna(
    log_numerical_data['item_id'].map(item_brand_map)
)
```

### Segment users by behavior

```python
def phan_nhom(row):
    if row['num_add_to_cart'] > 0 and row['num_purchase'] == 0:
        return ('Cart abandoner')
    elif row['num_add_to_cart'] > 0 and row['num_purchase'] > 0:
        return 'Cart to purchase'
    elif row['num_add_to_cart'] == 0 and row['num_purchase'] > 0:
        return 'Direct buyer'
    else:
   
        return 'No interaction'

user_behavior_summary['behavior_group'] = user_behavior_summary.apply(phan_nhom, axis=1)
user_behavior_summary
```

## COHORT RETENTION 

```python
log_data_processed['date_value'] = pd.to_datetime(log_data_processed['date_value'])
```

```python
# Cohort calculation
log_data_processed['date'] = pd.to_datetime(log_data_processed['date_value'])
log_data_processed['first_week'] = log_data_processed.groupby('user_id')['date'].transform('min').dt.to_period('W').dt.start_time
log_data_processed['activity_week'] = log_data_processed['date'].dt.to_period('W').dt.start_time
log_data_processed['cohort_week_index'] = ((log_data_processed['activity_week'] - log_data_processed['first_week'])
                                         / pd.Timedelta(weeks=1)).astype(int)
log_data_processed = log_data_processed[log_data_processed['cohort_week_index'] >= 0]  # Remove negative index

cohort_data = log_data_processed.groupby(['first_week', 'cohort_week_index'])['user_id'].nunique().reset_index()
cohort_pivot = cohort_data.pivot(index='first_week', columns='cohort_week_index', values='user_id')
cohort_pivot_filtered = cohort_pivot.dropna(thresh=4)

# Retention rate
cohort_sizes = cohort_pivot_filtered.iloc[:, 0].replace(0, 1)
retention = cohort_pivot_filtered.divide(cohort_sizes, axis=0)
```

```python
def plot_cohort_retention(retention_plot):

    #Visualize retention heatmap

    # Only show the day and month
    retention.index = retention.index.strftime('%d-%m')
    plt.figure(figsize=(18,12))

    # Heatmap plot
    ax = sns.heatmap(
        retention_plot,
        annot=True,
        fmt='.0%',  
        cmap='YlGnBu',
        linewidths=0.3,
        linecolor='white',
        cbar_kws={'format': '%.0f%%'},  # Color define 
        vmin=0,  
        vmax=1   
    )

    plt.title('Weekly Cohort Retention Heatmap', pad=20)
    plt.xlabel('Weeks from first online (cohort_week_index)')
    plt.ylabel('Cohort first_week')

    plt.tight_layout()
    plt.show()


plot_cohort_retention(retention)
```

**Remarks**:
- Cliff at Week 1: ~40–60% of users return to purchase the following week → need to improve cross-sell/upsell or send vouchers right after the first order
- Gradual decline until Week 4: down to ~20–40% → this is the stage where users need to be “nurtured” through email marketing, push notifications, or loyalty programs
- Plateau after Week 5: stabilizes around 20–40% up to Week 12+ → this group is the loyal customers (worth investing in for long-term retention)


## PREDICTING RETURN RATE OF CUSTOMERS

In this section, we will find the return rate of customers based on provided user_log dataset and take a quick look at models using the train and test datasets

### **PROPABILITY**

```python
# Sort the customers that actually purchase 
df_purchaser = log_data_processed[log_data_processed['action'] == 'purchase'].copy()
df_purchaser.shape
```

```python
df_purchaser['date_value'] = pd.to_datetime(df_purchaser['date_value'])

# Using polars to handle large datasets
import polars as pl

df = pl.from_pandas(df_purchaser)

# Rename to prepare self-join
df1 = df.rename({"date_value": "date1", "merchant_id": "merchant1"})
df2 = df.rename({"date_value": "date2", "merchant_id": "merchant2"})

# Self-join by user_id
df_joined = df1.join(df2, on="user_id", how="inner")

# Filter: merchant, date2 > date1, and <= date1 + 180 days (6 months)
df_filtered = df_joined.filter(
    (pl.col("merchant1") == pl.col("merchant2")) &
    (pl.col("date2") > pl.col("date1")) &
    (pl.col("date2") <= pl.col("date1") + pl.duration(days=180))
)

# Get unique user_id
users_returned = df_filtered.select("user_id").unique()

# Return rate calculation
return_rate = users_returned.height / df.select("user_id").unique().height
print(f"Return rate of customers in 6 month to the same merchant: {return_rate:.2%}")
```
